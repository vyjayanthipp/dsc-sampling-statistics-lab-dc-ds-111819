```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(2019)
mu, sigma = 10, 1.5
n = 1000

x_norm = np.random.normal(mu, sigma, n)
sns.distplot(x_norm);
plt.title("x_norm_Histogram")
plt.show()
#Create a variable called means and store the mean obtained from 1000 random samples.
means =[]
for sample in range(1000):
    means.append(np.random.normal(10, 1.5, 1000).mean())
sns.distplot(means) 
plt.title("means_histogram")
plt.show()

```
